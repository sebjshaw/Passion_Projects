# ATP_Matches_2021 Tennis Data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('data/atp_matches_2021.csv')
```

## Sorting the tourney_date column 
1. Turn into string 
2. Turn into datetime 
3. Sort the dataframe by date 
4. Reset the indices

```python
# converting the date into a datetime data type 

df = df.astype({'tourney_date':'string'})
df.tourney_date = pd.to_datetime(df.tourney_date)

df = df.sort_values(by =['tourney_date', 'match_num'])
df = df.reset_index()
```

## Data Analysis

```python
# who won the most games
df_games_won = pd.DataFrame(df.groupby('winner_name')['winner_name'].count()) 
df_games_won = df_games_won.rename(columns={'winner_name':'games_won'})
```

```python
# who lost the most games
df_games_lost = pd.DataFrame(df.groupby('loser_name')['loser_name'].count()) 
df_games_lost = df_games_lost.rename(columns={'loser_name':'games_lost'})
df_games_lost
```

```python
# who won the most tournaments
df[df['round'] == 'F']['winner_name'].value_counts().head()
```

```python
#find total games by first getting all the players and counting their occurrence in either winner_name or loser_name
all_players = set(list(df['winner_name'].unique()) + list(df['loser_name'].unique()))
total_games = []
for name in all_players:
    total_games.append([name , df[(df.winner_name == name) | (df.loser_name == name)]['index'].count()])

df_total_games = pd.DataFrame(total_games, columns=['player_name', 'total_games'])
df_total_games =df_total_games.set_index('player_name')
```

```python
# create dataframe of players, total games and games_won 
df_total_games = df_total_games.join(df_games_won).join(df_games_lost).sort_values(by=['total_games'], ascending=False)
```

```python
#replace n/a values with 0
df_players = df_total_games.replace(np.NaN, 0)
```

```python
# calculate win percentage
df_players['win_percent'] = (df_players.games_won / df_players.total_games) * 100
```

```python
# Who had the highest win percentage 
df_players = df_players.sort_values(by=['win_percent'], ascending=False)
df_players[df_players.total_games > 25].head()
```
